# Types of random graphs

```python
%matplotlib inline
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

## [Barabási–Albert model](https://en.wikipedia.org/wiki/Barab%C3%A1si%E2%80%93Albert_model)

```python
fig, ax = plt.subplots(8, 4, figsize=(10, 20))
for i in range(ax[0].size):
    g = nx.barabasi_albert_graph(5, 2, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[0,i])
    ax[0, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
for i in range(ax[1].size):
    g = nx.barabasi_albert_graph(10, 2, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[1,i])
    ax[1, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
for i in range(ax[2].size):
    g = nx.barabasi_albert_graph(15, 2, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[2,i])
    ax[2, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
for i in range(ax[3].size):
    g = nx.barabasi_albert_graph(20, 2, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[3,i])
    ax[3, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
for i in range(ax[4].size):
    g = nx.barabasi_albert_graph(5, 4, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[4,i])
    ax[4, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
for i in range(ax[5].size):
    g = nx.barabasi_albert_graph(10, 4, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[5,i])
    ax[5, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
for i in range(ax[6].size):
    g = nx.barabasi_albert_graph(15, 4, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[6,i])
    ax[6, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
for i in range(ax[7].size):
    g = nx.barabasi_albert_graph(20, 4, seed=i)
    nx.draw(g, node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[7,i])
    ax[7, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g):.2f}')
```

## [Watts–Strogatz model](https://en.wikipedia.org/wiki/Watts%E2%80%93Strogatz_model)

```python
fig, ax = plt.subplots(8, 4, figsize=(10, 20))
for i in range(ax[0].size):
    g = nx.watts_strogatz_graph(15, 2, p=.3, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[0,i])
    ax[0, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
for i in range(ax[1].size):
    g = nx.watts_strogatz_graph(15, 2, p=.8, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[1,i])
    ax[1, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
for i in range(ax[2].size):
    g = nx.watts_strogatz_graph(15, 4, p=.3, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[2,i])
    ax[2, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
for i in range(ax[3].size):
    g = nx.watts_strogatz_graph(15, 4, p=.8, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[3,i])
    ax[3, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
for i in range(ax[4].size):
    g = nx.watts_strogatz_graph(30, 2, p=.3, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[4,i])
    ax[4, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
for i in range(ax[5].size):
    g = nx.watts_strogatz_graph(30, 2, p=.8, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[5,i])
    ax[5, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
for i in range(ax[6].size):
    g = nx.watts_strogatz_graph(30, 4, p=.3, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[6,i])
    ax[6, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
for i in range(ax[7].size):
    g = nx.watts_strogatz_graph(30, 4, p=.8, seed=i)
    nx.draw(g,node_color=[d for n, d in g.degree()], node_shape='.', pos=nx.spring_layout(g, seed=i), ax=ax[7,i])
    ax[7, i].set_title(f'{nx.average_clustering(g):.2f} {nx.average_shortest_path_length(g) if nx.is_connected(g) else 0:.2f}')
```
